# Merge all 4 files 

```python
from functools import reduce
```

```python
dataframes = [barometer, indoor, outside, rainfall]

df_merged = reduce(lambda left,right: pandas.merge(left,right, on= 'index'), dataframes)

df_merged.head()
```

```python
#rename column headers

df_merged.columns = ['Date', 'Baro', 'Humidity', 'Indoor Temp', 'Indoor Temp Range Low', 'Indoor Temp Range High',
                'Outside Temp', 'Outside Temp Range Low', 'Outside Temp Range High', 'Rainfall mm' ]
```

# Monthly Values

```python
import numpy as np
```

```python
df_merged
```

```python
df_merged['month'] = pandas.DatetimeIndex(df_merged['Date']).month
# df_merged['month']
```

```python
m = df_merged.groupby([ "month"])
m
```

```python
monthly_ave_baro = m.aggregate({"Baro":np.mean})
monthly_ave_humidity = m.aggregate({"Humidity":np.mean})
monthly_ave_indoor = m.aggregate({"Indoor Temp":np.mean})
monthly_ave_indoor_low = m.aggregate({"Indoor Temp Range Low":np.mean})
monthly_ave_indoor_high = m.aggregate({"Indoor Temp Range High":np.mean})
monthly_ave_outside = m.aggregate({"Outside Temp":np.mean})
monthly_ave_outside_low = m.aggregate({"Outside Temp Range Low":np.mean})
monthly_ave_outside_high = m.aggregate({"Outside Temp Range High":np.mean})
monthly_ave_rainfall = m.aggregate({"Rainfall mm":np.mean})
```

```python
dataframes_ave = [monthly_ave_baro,
monthly_ave_humidity,
monthly_ave_indoor, 
monthly_ave_indoor_low, 
monthly_ave_indoor_high, 
monthly_ave_outside,
monthly_ave_outside_low, 
monthly_ave_outside_high, 
monthly_ave_rainfall]

df_merged_ave = reduce(lambda left,right: pandas.merge(left,right, on= 'month'), dataframes_ave)
```

```python
df_merged_ave.rename(index={1:'January',2:'February', 3:'March', 4:'April', 5:'May', 6:'June', 7:'July', 8:'August', 9:'September', 10:'October', 11:'November', 12:'December'}, inplace=True)
```

```python
df_merged_ave
```

# Seasonal Values


```python
df_merged2 = df_merged.copy()
```

```python
def seasonal(month):
    if month <= 2:
        return 1
    if month <= 5:
        return 2
    if month <= 8:
        return 3
    if month <= 11:
        return 4
    if month == 12:
        return 1
    
df_merged2['season'] = df_merged2['month'].apply(seasonal)
```

```python
s = df_merged2.groupby([ "season"])
```

```python
season_ave_baro = s.aggregate({"Baro":np.mean})
season_ave_humidity = s.aggregate({"Humidity":np.mean})
season_ave_indoor = s.aggregate({"Indoor Temp":np.mean})
season_ave_indoor_low = s.aggregate({"Indoor Temp Range Low":np.mean})
season_ave_indoor_high = s.aggregate({"Indoor Temp Range High":np.mean})
season_ave_outside = s.aggregate({"Outside Temp":np.mean})
season_ave_outside_low = s.aggregate({"Outside Temp Range Low":np.mean})
season_ave_outside_high = s.aggregate({"Outside Temp Range High":np.mean})
season_ave_rainfall = s.aggregate({"Rainfall mm":np.mean})
```

```python
dataframes_season_ave = [season_ave_baro,
season_ave_humidity,
season_ave_indoor, 
season_ave_indoor_low, 
season_ave_indoor_high, 
season_ave_outside,
season_ave_outside_low, 
season_ave_outside_high, 
season_ave_rainfall]

df_merged_season_ave = reduce(lambda left,right: pandas.merge(left,right, on= 'season'), dataframes_season_ave)
```

```python
df_merged_season_ave.rename(index={1:'Winter',2:'Spring', 3:'Summer', 4:'Fall'}, inplace=True)
```
